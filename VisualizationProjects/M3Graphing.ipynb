# Highlighting Important Graph Elements

This code shows how to use shading and text in a graph to highlight a specific element of a graph and draw the user's attention to the specific element.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Read in construction spending data using a pandas data frame

```python
dfconst = pd.read_csv('ConstructionTimeSeriesDataV2.csv')
dfconst.set_index(['Month'])
```

The code below graphs public and private construction spending over time on the plot and uses the ax.axvspan to highlight a specific period when the trend in spending seems to change as analyzed below.

```python
fig, ax = plt.subplots()
y = dfconst['Public Construction']
y2 = dfconst['Private Construction']
x = [str(i) for i in dfconst['Month-Year']]

ax.plot(x, y, color = 'b')
ax.plot(x, y2, color = 'darkorange')

ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction = 'in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction = 'in')
ax.set_ylabel('Spending (in billions)', fontsize = '14')
ax.set_xlabel('Date', fontsize = '14')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_ylim(min(y)-10000, max(y2) + 10000)
ax.axvspan('Sep-08', 'Jan-12', facecolor = 'k', alpha = 0.1)
ax.set_yticklabels(['${:1.0f}'.format(i/1000) for i in ax.get_yticks()])
ax.set_xticks(x[::12])
ax.set_xticklabels(x[::12])
ax.text('Mar-14', 45000,'Private',fontsize = '12', color = 'darkorange')
ax.text('Mar-14', 16000, 'Public', fontsize = '12', color = 'b')
ax.text('May-10', 70000, 'Great Recession', fontsize = 12, color = 'k', horizontalalignment = 'center')


fig.set_size_inches(14,6)
fig.suptitle('Did the Great Recession Impact Construction Spending?', fontsize='16')
```

There is a fairly obvious downturn in private construction spending during the Great Recession, starting around September 2008 and continuing through January 2012 when construction spending starts to increase again.  Public spending, however, remained fairly flat during the period, other than seasonal fluctuations.
